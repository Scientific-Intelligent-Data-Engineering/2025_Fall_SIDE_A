### 1. 문자열 "He has 12 apples and 5 oranges." 에서 숫자만 추출하는 정규 표현식을 작성하고, `search`, `findall`, `finditer` 세가지를 모두 사용하여 추출해보고 차이점을 설명하시오.

````python
text = "He has 12 apples and 5 oranges."

pattern = r"\d+"
m_search = re.search(pattern, text)
print("search:", m_search.group())
m_findall = re.findall(pattern, text)
print("findall:", m_findall)
print("finditer:")
for match in re.finditer(pattern, text):
    print(" - value:", match.group(), " position:", match.span())
````

### 2. 가장 많은 수를 가지고 있는 과일이 무엇인지 출력하시오. (힌트: grouping을 사용할 것)

````python
text = "He has 12 apples and 5 oranges."
pattern = r"(\d+)\s+(apples?|oranges?)"

fruit_dict = {}

for count, fruit in re.findall(pattern, text):
    count = int(count)
    fruit = fruit.rstrip('s')  #
    fruit_dict[fruit] = count

max_fruit = max(fruit_dict, key=fruit_dict.get)

print("과일별 개수:", fruit_dict)
print("가장 많은 과일:", max_fruit)
````

### 1. 이메일 주소가 유효한 이메일 형태로 작성되어 있는지 확인하는 함수 `is_valid_email`를 생성하고 유효한 이메일일 경우 이메일 주소를, 아닐 경우 `유효한 이메일 주소가 아닙니다.`를 출력하시오.

**Example**
```python
def is_valid_email(email):
  ...

address1 = 'test@gmail.com'
address2 = 'test.gmail.com'

print(is_valid_email(address1))
>> test@gmail.com
print(is_valid_email(address2))
>> 유효한 이메일 주소가 아닙니다.
```

````python
def is_valid_email(email):
  pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

  if re.match(pattern, email):
    return email
  else :
    return "유효한 이메일 주소가 아닙니다."
address1 = 'test@gmail.com'
address2 = 'test.gmail.com'

print(is_valid_email(address1))
print(is_valid_email(address2))
````

### 2. Q2.1번 문제에 이어서 유효한 이메일일 경우, 해당 이메일의 아이디와 도메인을 출력하시오.:

**Example**

```python
def is_valid_email2(email):
  ...

address1 = 'test@gmail.com'
address2 = 'test.gmail.com'

print(is_valid_email2(address1))
>> ID: test, Domain: gmail.com
print(is_valid_email2(address2))
>> 유효한 이메일 주소가 아닙니다.

```

````python

def is_valid_email2(email):
    pattern = r"^([\w\.-]+)@([\w\.-]+\.\w+)$"

    match = re.match(pattern, email)
    if match:
        user_id = match.group(1)
        domain = match.group(2)
        return f"ID: {user_id}, Domain: {domain}"
    else:
        return "유효한 이메일 주소가 아닙니다."
address1 = 'test@gmail.com'
address2 = 'test.gmail.com'

print(is_valid_email2(address1))
print(is_valid_email2(address2))
````

### 정규표현식을 사용해서 html의 문자열을 파싱하는 beautifulsoup의 find()와 get_text() function을 갖는 ParserWithRegex라는 클래스를 구현해보시오.

**Example**
```python
class ParserWithRegex:
    def __init__(self, html):
        self.source = html

    def find(self, tag):
      ...

    def get_text(self, html_fragment):
      ...


html = """
<html>
<body>
<div>Hello <b>World</b>!</div>
<p>This is a paragraph.</p>
<div class=test>Another <i>div</i>.</div>
</body>
</html>
"""

parser = ParserWithRegex(html)

div_content = parser.find('div')
print(div_content)
>> Hello <b>World</b>!

text = parser.get_text(div_content)
print(text)
>> Hello World!
```

````python

class ParserWithRegex:
    def __init__(self, html):
        self.source = html

    def find(self, tag):
        pattern = fr"<{tag}[^>]*>(.*?)</{tag}>"

        match = re.search(pattern, self.source, re.DOTALL)
        if match:
            return match.group(1).strip()
        else:
            return None

    def get_text(self, html_fragment):
        no_tags = re.sub(r"<[^>]+>", "", html_fragment)
        return no_tags.strip()

html = """
<html>
<body>
<div>Hello <b>World</b>!</div>
<p>This is a paragraph.</p>
<div class=test>Another <i>div</i>.</div>
</body>
</html>
"""

parser = ParserWithRegex(html)

div_content = parser.find('div')
print(div_content)

text = parser.get_text(div_content)
print(text)
````

```
    </ul>
</body>
</html>
```


#### 1. 페이지의 제목(`<title>`)을 추출하여라.

````python
from bs4 import BeautifulSoup

html = """
<!DOCTYPE html>
<html>
<head>
    <title>Sample Web Page</title>
</head>
<body>
    <h1>Welcome to the Sample Web Page</h1>
    <div id="author">
        Author: Hyeongwon Kang
    </div>
    <div id="content">
        <p>This is a paragraph in the web page.</p>
        <p>This is another paragraph, with a <a href="https://www.example.com">link</a>.</p>
    </div>
    <ul id="items">
        <li class="item">Item 1</li>
        <li class="item">Item 2</li>
        <li class="item">Item 3</li>
    </ul>
</body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")

# 1. 페이지의 제목(<title>) 추출
title = soup.title.get_text()

# 2. id="content" div 내부 모든 텍스트 추출
content_text = soup.find(id="content").get_text(strip=True)

# 3. id="items" ul 내부 모든 <li> 텍스트 리스트로 추출
items = [li.get_text(strip=True) for li in soup.find(id="items").find_all("li")]

print("1. 페이지 제목:", title)
print("2. content div 텍스트:", content_text)
print("3. items 리스트:", items)
````

#### 2. "content"라는 ID를 가진 `<div>` 태그의 모든 텍스트를 추출하여라.

#### 3. "items"라는 ID를 가진 `<ul>` 태그 내에 있는 모든 항목(`<li>`)의 텍스트를 리스트로 만들어 반환하여라.

### 다음 기사의 본문에서 강필성 교수님이 몇번 언급되는지 구하시오.

https://www.seoul.co.kr/news/newsView.php?id=20200108500123&wlog_tag3=naver

> 기사 본문을 크롤링하고 정규표현식을 사용합니다.
>
> '강 교수'도 강필성 교수님 언급으로 카운트 되어야 합니다.
>
> 패턴은 단 한번만 사용 가능합니다.

````python
import requests, re
from bs4 import BeautifulSoup

NEWS_URL = "https://www.seoul.co.kr/news/newsView.php?id=20200108500123"

request_headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept-Language": "ko-KR,ko;q=0.9",
    "Referer": "https://www.google.com"
}

response = requests.get(NEWS_URL, headers=request_headers)
response.raise_for_status()

response.encoding = "utf-8"

soup_obj = BeautifulSoup(response.text, "html.parser")

content_div = soup_obj.select_one("div.viewContent.body18.color700")

article_text = content_div.get_text(" ", strip=True)

keyword_pattern = re.compile(r"강(필성)?\s*교수")
mention_count = len(keyword_pattern.findall(article_text))

print("언급 횟수:", mention_count)
````

**try, except 설명**
- `try`와 `except`는 파이썬에서 예외 처리를 위해 사용되는 구문이다. 이 구문을 사용하면 코드 실행 중에 발생할 수 있는 오류(예외)를 처리하고, 프로그램이 예외로 인해 갑자기 중단되는 것을 방지할 수 있다. `try` 블록 안의 코드를 실행하다가 오류가 발생하면 즉시 그 블록을 중단하고 `except` 블록으로 이동한다.
```python
try:
    예외가 발생할 수 있는 코드
except:
    Error가 발생했을 때 실행될 코드
```
