1. 로또 입력 번호 자동으로 생성해주는 함수 `get_auto_lotto_numbers`를 구현해본다. 함수의 입력 값은 줄의 개수이며 1줄부터 5줄까지 생성할 수 있다. 더불어, seed는 설정되어서는 안되며, 결과값으로 나오는 번호들은 각 줄(행) 마다 오름차순이어야 한다.

  ```
  ex)
  def get_auto_lotto_numbers(n:int):
    if ...:
      ...
    else:
      raise BaseException(f"줄의 개수를 다시 입력하시오!")

  lotto_numbers = get_auto_lotto_numbers(n=5)
  display(lotto_numbers)
  >>>
  array([[ 2, 11, 15, 23, 29, 45],
        [ 5, 17, 25, 31, 35, 44],
        [ 5,  8, 10, 13, 21, 25],
        [ 4,  5, 14, 21, 26, 36],
        [ 2,  7,  9, 12, 14, 17]])
  ```


````python
def get_auto_lotto_numbers(n):
  if 1<=n and 5>=5:
    row = []
    for _ in range(n):
      num = np.random.choice(np.arange(1, 46), size=6, replace=False)
      num.sort()
      row.append(num)
    return np.array(row)
  else:
    raise BaseException(f"줄의 개수를 다시 입력하시오!")

lotto_numbers = get_auto_lotto_numbers(n=5)
display(lotto_numbers)
````

2. 당첨번호를 생성하는 함수 `get_winning_numbers`를 구현해본다. 1부터 45까지의 숫자 중 6개를 중복없이 무작위 추출한 후 오름차순으로 정렬된 당첨번호를 생성한다. 단, seed를 입력받을 수 있도록 한다 (seed는 실제 로또의 회차 번호라고 생각하면 된다).
```
winning_numbers = get_winning_numbers(seed=1092)
display(winning_numbers)
>>>
array([ 1,  4,  8, 15, 22, 37])
```

````python
def get_winning_numbers(seed):
  np.random.seed(seed)
  li = np.random.choice(np.arange(1, 46), size=6, replace=False)
  li.sort()
  return li

winning_numbers = get_winning_numbers(seed=1197)
display(winning_numbers)
````

3. 1.과 2.에서 각각 생성된 입력 번호와 당첨 번호를 비교하여 결과를 알려주는 `get_lotto_results`를 구현해본다.
입력 번호와 당첨 번호를 함수의 입력값으로 설정하여 각 줄 마다 몇개의 숫자가 당첨됐는지 알려주도록 출력한다.
```
get_lotto_results(lotto_numbers,winning_numbers)
>>>
<< 당첨번호: [ 1  4  8 15 22 37] >>
1 번째 줄 당첨된 번호 개수: 0
2 번째 줄 당첨된 번호 개수: 1
3 번째 줄 당첨된 번호 개수: 0
4 번째 줄 당첨된 번호 개수: 0
5 번째 줄 당첨된 번호 개수: 0

```

````python
def get_lotto_results(lotto_numbers, winning_numbers):
  print(f"<< 당첨번호: {winning_numbers.tolist()} >>")

  for i, row in enumerate(lotto_numbers, start=1):
    matched = np.isin(row, winning_numbers)

    cnt = matched.sum()
    print(f"{i} 번째 줄 당첨된 번호 개수: {cnt}")

get_lotto_results(lotto_numbers, winning_numbers)
````

4. 다음은 로또 결과를 화면으로 확인했을 때의 모습의 예시다.  
![로또 QR코드 확인](https://blog.kakaocdn.net/dn/wCnKq/btqFvuVPxpz/wrzMfInz4wJD9eB0JeC78K/img.png)

이처럼 입력 번호 중 당첨 번호에 포함 된 것을 표시할 수 있는 함수인 `get_lotto_display_results`를 구현해본다.  
이때, 당첨 번호에 해당하는 번호는 그대로 표시하고 미당첨 번호에 해당하는 것은 'x'로 표시하도록 한다.
```
get_lotto_display_results(lotto_numbers,winning_numbers)
>>>
[['x' 'x' 'x' 'x' 'x' 'x']
 ['x' '15' 'x' 'x' 'x' 'x']
 ['x' 'x' 'x' 'x' '37' 'x']
 ['1' 'x' 'x' 'x' 'x' 'x']
 ['x' 'x' 'x' 'x' 'x' 'x']]
```

````python
def get_lotto_display_results(lotto_numbers, winning_numbers):
  display_result = []

  for row in lotto_numbers:
    flag = np.isin(row, winning_numbers)

    shown_row = []

    for num, matched in zip(row, flag):
      if matched:
        shown_row.append(str(num))
      else:
        shown_row.append('x')

    display_result.append(shown_row)

  return display_result

get_lotto_display_results(lotto_numbers, winning_numbers)
````

* 중복된 내역은 마지막 하나만 남겨두고 다 제거한다.  

위의 규칙으로 응모 접수 데이터를 전처리하는 함수 `preprocess`를 생성하고 실제 결과를 출력하시오.(`NikeDraw.csv` 파일의 encoding은 'cp949'로 설정하여야 함)

````python
path = "/content/drive/MyDrive/Colab Notebooks/위티 데이터분석 세션/NikeDraw.csv"
data_df = pd.read_csv(path, encoding="cp949")

display(data_df)

def preprocess(df):
  temp = df.copy()

  must_cols = ['이메일', '사이즈', '응모지점']
  temp = temp.dropna(subset=must_cols)

  temp['응모횟수'] = temp.groupby('이메일').cumcount()
  temp = temp[temp['응모횟수'] < 2]
  temp = temp.drop(columns='응모횟수')

  temp = temp.drop_duplicates(subset=['이메일', '사이즈', '응모지점'], keep='last')

  return temp

final_df = preprocess(data_df)
display(final_df)
````

2. 1에서 전처리 완료된 최종 응모 내역 데이터 `final_df`에서 50%가 무작위로 당첨된다고 하자. 이를 구현하는 함수 `get_results`를 생성하고 `final_df`를 입력한 결과를 출력하시오.

  * 50%의 의미: 최종 응모내역 10건 → 5건 당첨 / 최종 응모내역 11건 → 5건 당첨
  * seed나 random_state는 별도로 지정하지 않음

    ```
    ex)
    get_results(final_df)
    >>>
              이메일	사이즈	응모지점
    14	gunho_no@korea.ac.kr	270.0	안암점
    0	doyooni303@korea.ac.kr	265.0	목동점
    1	gunho_no@korea.ac.kr	270.0	목동점
    13	seonghee_hong@korea.ac.kr	275.0	영등포점
    6	sangmin_lee@korea.ac.kr	270.0	안암점
    ```


````python
def get_results(final_df):
  total = len(final_df)
  win_n = total // 2
  winners = final_df.sample(n=win_n, replace=False)
  return winners

winners_df = get_results(final_df)
display(winners_df)
````
