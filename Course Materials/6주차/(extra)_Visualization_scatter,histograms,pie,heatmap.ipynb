### 3 scatter plot

```python
%matplotlib ipympl
from mpl_toolkits.mplot3d import Axes3D

n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')

plt.show()
```

## Histograms, Binnings, and Density
-  (Histogram)    ,  ,   (  ) 

[Document](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
output.disable_custom_widget_manager()
%matplotlib inline
np.random.seed(1234)
data = np.random.randn(100)
print(f"data:\n{data}")

plt.hist(data)
plt.show()
```

####   
- hist()  bins      

```python
plt.hist(data, bins=20) # default: 10
plt.show()
```

####   
- cumulative  True    

```python
plt.hist(data, cumulative=True, label='cumulative=True')
plt.hist(data, cumulative=False, label='cumulative=False')
plt.legend(loc='upper left')
plt.show()
```

####   
- histtype     {bar, barstacked, step, stepfilled}    ,  bar

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='step', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.show()
```

## Pie chart
-   (Pie chart,  )      
-     ratio labels , pie()   
- autopct      

[Document](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
ratio = [15, 30, 45, 10]

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

####    
- tartangle     
-  0 (  x) 
- counterclock=False       

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
ratio = [15, 30, 45, 10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
```

####    
- explode       

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
ratio = [15, 30, 45, 10]
explode = [0, 0.10, 0.10, 0]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

####  
shadow True ,    

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
ratio = [15, 30, 45, 10]
explode = [0.05, 0.05, 0.05, 0.05]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True)
plt.show()
```

####  

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
ratio = [15, 30, 45, 10]
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['#FF6B6B', '#00B894', '#6C5CE7', '#FFD32A']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

####   
- wedgeprops    

```python
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
ratio = [15, 30, 45, 10]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 1, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

```python
import seaborn as sns

tips_df = sns.load_dataset('tips')
tips_df.rename(columns = {'sex':'gender'},inplace = True)
display(tips_df)
```

Q.1       ?

```python
def get_ratio(x):
  return x/sum(x)
gender_counts=tips_df['gender'].value_counts().to_frame()
gender_counts['ratio']=gender_counts.apply(get_ratio,axis=0);gender_counts

wedgeprops={'width': 1, 'edgecolor': 'w', 'linewidth': 5}
plt.pie(gender_counts['ratio'], labels=gender_counts.index, autopct='%.2f%%', counterclock=False, wedgeprops=wedgeprops)
plt.show()
```

```python
#  pie method  x/sum(x)  .! 
plt.pie(gender_counts['gender'], labels=gender_counts.index, autopct='%.2f%%', counterclock=False, wedgeprops=wedgeprops)
plt.show()
```

Q.2        ?

```python
day_bills = tips_df.groupby('day')['total_bill'].mean();day_bills
day_bills_dict = day_bills.to_dict();day_bills_dict
labels = list(day_bills_dict.keys())
ratio = list(day_bills_dict.values())
plt.pie(ratio, labels = labels, autopct=r'%.2f%%', counterclock=False, wedgeprops=wedgeprops)
plt.title("Total bill ratio by day")
plt.show()
```

*** Pandas  plot**
---
[[ ]](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)  
* kind :The kind of plot to produce  

  line : line plot (default)

  bar : vertical bar plot

  barh : horizontal bar plot

  hist : histogram

  box : boxplot

  kde : Kernel Density Estimation plot

  density : same as kde

  area : area plot

  pie : pie plot

  scatter : scatter plot (DataFrame only)

  hexbin : hexbin plot (DataFrame only)

```python
pie_kwargs = {
    'kind':'pie',
    'autopct':r'%.2f%%',
    'wedgeprops':{'width': 1, 'edgecolor': 'w', 'linewidth': 5},
    'title':"Pie Chart(Gender ratio)"
}
plt.figure(figsize=(5,5))
tips_df['gender'].value_counts().plot(**pie_kwargs)
plt.show()
```

```python
hist_kwargs=dict(
    kind='hist',
    alpha=0.3,
    bins=20,
)
tips_df.groupby('day')['total_bill'].plot(**hist_kwargs)
plt.legend()
plt.show()
```

```python
plt.matshow(data)
plt.colorbar(shrink=0.8, aspect=10)
plt.show()
```

```python
plt.matshow(data)
plt.colorbar(shrink=0.8, aspect=10) # shrink = , aspect = 
plt.clim(0.2, 0.8)
plt.show()
```
