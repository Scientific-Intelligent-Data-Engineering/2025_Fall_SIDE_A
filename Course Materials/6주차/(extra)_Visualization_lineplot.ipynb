## 0. matplotlib

- Matplotlib은 NumPy 배열을 기반으로 구축된 다중 플랫폼 데이터 시각화 라이브러리

- Matplotlib을 사용하면 다양한 유형의 플롯을 만들 수 있으며, 이는 데이터 분석 및 시각적 표현에 매우 유용

#### linestyle 바꾸기

- 선 종류를 나타내는 문자열 또는 튜플을 이용해서 다양한 선의 종류를 구현할 수 있다.

<p align="center"> <img src="https://wikidocs.net/images/page/137791/set_linestyle_05.png" width=33%><img src="https://wikidocs.net/images/page/137791/set_linestyle_06.png" width=33%> <img src="https://wikidocs.net/images/page/137791/set_linestyle_07.png" width=33%></p>


```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':')  # dotted

plt.plot(x, x + 8, linestyle=(0, (1, 0)))  # solid
plt.plot(x, x + 9, linestyle=(0, (5, 1))) # dashed
plt.plot(x, x + 10, linestyle=(0, (3, 5, 1, 5))) # dashdot
plt.plot(x, x + 11, linestyle=(0, (1, 1)))  # dotted
plt.show()
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r')  # dotted red
plt.show()
```

### 축 범위 지정하기

그래프의 X, Y축이 표시되는 범위를 지정

- `xlim()` - X축이 표시되는 범위를 지정하거나 반환
- `ylim()` - Y축이 표시되는 범위를 지정하거나 반환
- `axis()` - X, Y축이 표시되는 범위를 지정하거나 반환

#### xlim(), ylim()

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
plt.show()
```

#### axis()

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
plt.show()
```

#### 옵션
'on' | 'off' | 'equal' | 'scaled' | 'tight' | 'auto' | 'normal' | 'image' | 'square'

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
plt.show()
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
plt.show()
```

#### 축 범위 얻기

- xlim(), ylim() 함수는 그래프 영역에 표시되는 X축, Y축의 범위를 각각 반환

- axis() 함수는 그래프 영역에 표시되는 X, Y축의 범위를 반환

```python
plt.plot(x, np.sin(x))
x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('scaled')
print(axis_range)
plt.show()
```

### 범례 표시하기

- 범례 (Legend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트

- legend() 함수를 사용해서 그래프에 범례를 표시할 수 있음

- Doc: [링크](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
plt.show()
```

#### 위치 지정
legend() 함수의 loc 파라미터를 이용해서 범례가 표시될 위치를 설정할 수 있음

loc 파라미터를 숫자 쌍 튜플로 지정하면, 해당하는 위치에 범례가 표시

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend(loc=(0.5, 0.5))
plt.show()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend(loc='lower right')
plt.show()
```

#### 열 개수 지정

legend() 함수의 ncol 파라미터는 범례에 표시될 텍스트의 열의 개수를 지정

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend(loc='best', ncol=2)
plt.show()
```

#### 폰트 크기

legend() 함수의 fontsize 파라미터는 범례에 표시될 폰트의 크기를 지정

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend(loc='best', ncol=2, fontsize=14)
plt.show()
```

#### 범례 테두리

- frameon 파라미터는 범례 텍스트 상자의 테두리를 표시할지 여부를 지정

- shadow 파라미터를 사용해서 텍스트 상자에 그림자를 표시 가능

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend(loc='best', ncol=2, frameon=True, shadow=True)
plt.show()
```

### 타이틀 설정

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot')
plt.show()
```

- plt.title() 함수의 loc 파라미터를 {‘left’, ‘center’, ‘right’} 중 선택할 수 있으며 디폴트는 ‘center’

- pad 파라미터는 타이틀과 그래프와의 간격을 포인트 단위로 설정

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve", loc='right', pad=20)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

### 그리드 설정
- 데이터의 위치를 더 명확하게 나타내기 위해 그래프에 그리드 (Grid, 격자)를 표시할 수 있음

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```

#### 축 지정
- axis 값으로 {‘both’, ‘x’, ‘y’} 중 선택할 수 있고 디폴트는 ‘both’

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True, axis='y')
plt.show()
```

#### 스타일 바꾸기
- color, alpha, linestyle 파마리터를 사용해서 그리드 선의 스타일을 설정 가능

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True, axis='y', color='red', alpha=0.5, linestyle='--')
plt.show()
```

### 수평선/수직선 그리기

#### 수평선
- axhline()
  - 첫번째 인자는 y 값으로서 수평선의 위치
  - 두, 세번째 인자는 xmin, xmax 값으로서 0에서 1 사이의 값을 입력(0은 왼쪽 끝, 1은 오른쪽 끝)

- hlines() 함수에 y, xmin, xmax를 순서대로 입력하면, 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.axhline(1.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 4.0, 6.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

#### 수직선
- axvline()
  - 첫번째 인자는 x 값으로서 수직선의 위치
  - 두, 세번째 인자는 ymin, ymax 값으로서 0에서 1 사이의 값을 입력(0은 아래쪽 끝, 1은 위쪽 끝)

- vlines() 함수에 x, ymin, ymax를 순서대로 입력하면, 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시


```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```
